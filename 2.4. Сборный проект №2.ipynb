# Технологический процесс
Как золото получают из руды? Изучите внимательно стадии процесса.
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.
 
Опишем каждую стадию: 
1. Флотация
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

2. Очистка 
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.
Описание данных
Технологический процесс
- `Rougher feed` — исходное сырье
- `Rougher additions` (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant 
- `Xanthate` — ксантогенат (промотер, или активатор флотации);
- `Sulphate` — сульфат (на данном производстве сульфид натрия);
- `Depressant` — депрессант (силикат натрия).
- `Rougher process` (англ. «грубый процесс») — флотация
- `Rougher tails` — отвальные хвосты
- `Float banks` — флотационная установка
- `Cleaner process` — очистка
- `Rougher Au` — черновой концентрат золота
- `Final Au` — финальный концентрат золота
## Параметры этапов
- `air amount` — объём воздуха
- `fluid levels` — уровень жидкости
- `feed size` — размер гранул сырья
- `feed rate` — скорость подачи
## Наименование признаков

Наименование признаков должно быть такое:

**`[этап].[тип_параметра].[название_параметра]`**

Пример: `rougher.input.feed_ag`

Возможные значения для блока `[этап]`:
- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики

Возможные значения для блока `[тип_параметра]`:
- `input` — параметры сырья
- `output` — параметры продукта
- `state` — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики

Для расчета эффективности обогащения нам потребуются следующие данные:
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

```python
C = train['rougher.output.concentrate_au'] # доля золота в концентрате после флотации/очистки
F = train['rougher.input.feed_au'] # доля золота в сырье/концентрате до флотации/очистки
T = train['rougher.output.tail_au'] # доля золота в отвальных хвостах после флотации/очистки
```

Теперь рассчитаем эффективность по имеющейся формуле и рассчитаем `MAE`

```python
calculated_recovery = C*(F-T)*100/(F*(C-T))
recovery = train['rougher.output.recovery']
MAE = mean_absolute_error(recovery, calculated_recovery)
print(f'Mean absolute error эффективности обогащения: {MAE}')
```

### Функция sMAPE
Для начала напишем функцию для sMAPE, потом уже для итоговой sMAPE

```python
def smape_(tar, pred): # tar = target, pred = prediction
    return (abs(pred - tar) / ((abs(tar) + abs(pred)) / 2)).sum() / len(tar)
scorer = make_scorer(smape_)
def final_smape_(model, features_train = features_train, target_rougher_train = target_rougher_train, target_final_train = target_final_train):
    # кросс-валидация с отбором значений sMAPE
    cross_valid_score_rougher = cross_val_score(model, features_train, target_rougher_train, scoring = scorer, cv = 5)
    cross_valid_score_final = cross_val_score(model, features_train, target_final_train, scoring = scorer, cv = 5)
    # расчеты 
    cross_valid_score_rougher_mean = cross_valid_score_rougher.mean()
    cross_valid_score_final_mean = cross_valid_score_final.mean()
    final = 0.25 * cross_valid_score_rougher_mean + 0.75 * cross_valid_score_final_mean
    return final
```

Написали необходимую функцию, теперь можно приступать к обучению моделей

#### Решающее дерево
Проверим модель регрессии дерева решений

```python
max_depth_DTR = 1
final_smape = 100
for depth in range(1, 15):    
    DTR_model = DecisionTreeRegressor(max_depth = depth, random_state = 12345)
    smape = final_smape_(DTR_model)
    if smape <= final_smape:
        final_smape = smape
        best_depth_DTR = depth
        best_model_DTR = DTR_model
```

```python
print(f'Лучшее значение final sMAPE для регрессионного дерева решений: {final_smape:.3%}, достигнутый при глубине {best_depth_DTR}')
```

#### Случайный лес
Теперь проверим модель случайного леса. Начнем с подбора необходимой глубины на малом количестве деревьев, затем рассмотрим большее их количество

```python
max_depth_RFR = 1
final_smape = 100
for depth in range(1,15):    
    RFR_model = RandomForestRegressor(max_depth = depth, n_estimators=10, random_state = 12345)
    smape = final_smape_(RFR_model)
    if smape <= final_smape:
        final_smape = smape
        best_depth_RFR = depth
```

```python
print(f'Лучшая глубина случайного леса на 10 деревьях: {best_depth_RFR}')
```

Получили значение самой лучшей глубины, теперь необходимо выяснить лучшее значение количества деревьев для данной глубины

```python
for est in range(1,50,5):    
    RFR_model = RandomForestRegressor(max_depth = best_depth_RFR, n_estimators=est, random_state = 12345)
    smape = final_smape_(RFR_model)
    if smape <= final_smape:
        final_smape = smape
        best_n_est_RFR = est
        best_model_RFR = RFR_model
```

```python
print(f'Лучшее значение final sMAPE для случайного леса: {final_smape:.3%}, достигнутое на глубине {best_depth_RFR} и при количестве деревьев: {best_n_est_RFR}')
```
