## Обучение

### Создание признаков в датафрейме
Напишем функцию для создания признаков в датафрейме

Введем тригонометрические преобразователи

```python
def sin_transformer(period):
    return FunctionTransformer(lambda x: np.sin(x / period * 2 * np.pi))


def cos_transformer(period):
    return FunctionTransformer(lambda x: np.cos(x / period * 2 * np.pi))
```

```python
def make_features(df, max_lag, rolling_mean_size, plot_needed = False):
    data = df.copy()
    data['hour'] = data.index.hour
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour_sin'] = sin_transformer(24).fit_transform(data)['hour']
    data['hour_cos'] = cos_transformer(24).fit_transform(data)['hour']
    data['dayofweek_sin'] = sin_transformer(7).fit_transform(data)['dayofweek']
    data['dayofweek_cos'] = cos_transformer(7).fit_transform(data)['dayofweek']
    
    if plot_needed == True:   
        data[['hour', 'hour_sin', 'hour_cos']].head(24).plot(x="hour")
        plt.title("Trigonometric encoding for the 'hour' feature")
        plt.show()

        fig, ax = plt.subplots(figsize=(7, 5))
        sp = ax.scatter(data["hour_sin"], data["hour_cos"], c=data["hour"])
        ax.set(
            xlabel="sin(hour)",
            ylabel="cos(hour)",
        )
        _ = fig.colorbar(sp)
        plt.show()
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    return data
```

Функция для расчета `RMSE` кросс-валидацией

```python
def evaluate(model, features, target, cv):
    cv_results = cross_val_score(
        model,
        features,
        target,
        cv=cv,
        scoring='neg_mean_squared_error',
    )
    rmse = (-cv_results)**0.5
    print(
        f"Root Mean Squared Error: {rmse.mean():.3f} +/- {rmse.std():.3f}"
    )
    return rmse.mean(), rmse.std()
```

### Преобразование данных
Напишем функцию, которая выделит признаки и целевой признак, а также разобьёт данные на обучающую и тестовую выборки в соотношении 9:1

До тестирования будем использовать кросс-валидацию

```python
df = df.drop(['mean', 'std'], axis = 1)
```

```python
def df_split(df):   
    train, test = train_test_split(df, shuffle=False, test_size=0.1)
    train = train.dropna()
    features_train = train.drop('num_orders', axis = 1)
    features_test = test.drop('num_orders', axis = 1)
    target_train = train['num_orders']
    target_test = test['num_orders']    
    return features_train, features_test, target_train, target_test
```

У нас есть функция для добавления признаков в датафрейм, но неизвестно, какие параметры к ней применять
Для того, чтобы это сделать, посмотрим динамику RMSE на линейной регрессии

```python
# сначала определим точку опоры, относительно которой будем смотреть получающиеся rmse
df_new = make_features(df, 1, 1, True)
features_train, features_test, target_train, target_test = df_split(df_new)

ts_cv = TimeSeriesSplit(n_splits=5) 

model = LinearRegression()
best_rmse, std = evaluate(model, features_train, target_train, ts_cv)
```

```python
for max_lag in [24, 48, 72, 96, 120, 144, 168]:
    for rolling_mean_size in [24, 48, 72, 96, 120, 144, 168]:
        df_new = make_features(df, max_lag, rolling_mean_size)
        features_train, features_test, target_train, target_test = df_split(df_new)
        model = LinearRegression()
        rmse, std = evaluate(model, features_train, target_train, ts_cv) 
        if rmse <= best_rmse:
            best_rmse = rmse
            best_lag = max_lag
            best_roll = rolling_mean_size
```

```python
print(f'Лучший RMSE: {best_rmse:.3f}\nЛучшие отстающие значения: {best_lag}\nЛучшее скользящее среднее: {best_roll}')
```
