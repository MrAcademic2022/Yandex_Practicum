PS: сначала проведем все манипуляции с несбалансированными моделями, а после них подведём промежуточный итог, где поднимем все получившиеся значения метрик

Напишем функцию, которая будет выдавать нам основные метрики по вводным данным

```python
def metrics(target, prediction, predict_proba, model_name = 'unknown', model_type = 'unknown'):
    # основные метрики
    print(f'{"Accuracy:":<10}',accuracy_score(target, prediction),\
        '\n' + f'{"Recall:":<10}', recall_score(target, prediction), \
        '\n' + f'{"Precision:":<10}', precision_score(target, prediction), \
        '\n' + f'{"F1-score:":<10}', f1_score(target, prediction),\
        '\n'+ f'{"AUC-ROC:":<10}', roc_auc_score(target, predict_proba))
    
    # матрица
    ConfusionMatrixDisplay(confusion_matrix(target, prediction)).plot()
    plt.title(f'Confusion matrix модели \n{model_name}\nСоотношение классов: {model_type}')
    plt.show()
    
    # precision-recall curve
    precision, recall, thresholds = precision_recall_curve(target, predict_proba)

    plt.figure(figsize=(6, 6))
    plt.step(recall, precision, where='post')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title(f'Кривая Precision-Recall модели \n{model_name}\nСоотношение классов: {model_type}')
    plt.show() 

    #roc-curve
    fpr, tpr, thresholds = roc_curve(target, predict_proba) 
    plt.figure() 
    plt.plot(fpr,tpr)
    # ROC-кривая случайной модели
    plt.plot([0, 1], [0, 1], linestyle='--')
    #   установим границы осей от 0 до 1 
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    #   подпишем оси "False Positive Rate" и "True Positive Rate"
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    # добавим заголовок "ROC-кривая" функцией plt.title() 
    plt.title(f"ROC-кривая модели \n{model_name}\nСоотношение классов: {model_type}")
    plt.show()

    return {"type":model_type,
            "accuracy":accuracy_score(target, prediction),
            "recall":recall_score(target, prediction),
            "precision":precision_score(target, prediction),
            "f1-score":f1_score(target, prediction),
            "auc_roc":roc_auc_score(target, predict_proba)}
    
```

Функция, которая будет выдавать нам результат исследования модели случайного леса (лучшую глубину при максимизации f1-score, все метрики качества, различные графики) для данных условий (определнных обучающих и валидационных выборок, наличии/отсутствии баланса классов)

```python
def decision_tree_research(features_train, 
                           target_train, 
                           features_valid, 
                           target_valid, 
                           class_weight = None,
                           type = 'unknown'):   
    # ищем лучший f1-score, запоминаем, какая глубина ему соответствует, параллельно
    # записываем все значения f1-score для разных глубин в массив, чтобы потом 
    # вывести в виде графика на экран, а также запоминаем лучшую модель (по f1-score)
    f1_best = 0
    f1_score_list = []
    dec_tree_best = 'None'
    for depth in range(1,20):
        model = DecisionTreeClassifier(max_depth = depth, 
                                    class_weight=class_weight,
                                    random_state=12345)
        model.fit(features_train, target_train)
        f1 = f1_score(target_valid, model.predict(features_valid))
        f1_score_list.append(f1)
        if f1 >= f1_best:
            f1_best = f1
            best_depth = depth
            dec_tree_best = model
    plt.plot(f1_score_list)
    plt.title(f'Зависимость f1-score решающего дерева от глубины')
    plt.xlabel('Глубина')
    plt.ylabel('f1-score')
    plt.show()
    
    print(f'Лучший f1-score для решающего дерева: {f1_best}\n\
          Лучшая глубина: {best_depth}\n')
    # возвращаем лучшую по метрике f1-меры модель решающего дерева и его основные метрики
    return dec_tree_best, metrics(target_valid, 
        dec_tree_best.predict(features_valid),
        dec_tree_best.predict_proba(features_valid)[:,1],
        'DecisionTreeClassifier',
        type)
    
```

Функция, которая будет выдавать нам результат исследования модели случайного леса (лучшие глубину и количество деревьев при максимизации f1-score, все метрики качества, различные графики) для данных условий (определнных обучающих и валидационных выборок, наличии/отсутствии баланса классов)

```python
def random_forest_research(features_train, 
                           target_train, 
                           features_valid, 
                           target_valid, 
                           class_weight = None,
                           type = 'unknown'):  
    # ищем лучший f1-score для 1-20 деревьев, запоминаем, какая глубина ему соответствует
    f1_best = 0
    f1_score_list = []
    for depth in range(1,20):
        for n_est in range(1,20):    
            model = RandomForestClassifier(max_depth = depth, 
                                        n_estimators= n_est, 
                                        class_weight=class_weight,
                                        random_state=12345)
            model.fit(features_train, target_train)
            f1 = f1_score(target_valid, model.predict(features_valid))
            #f1_score_list.append(f1)
            if f1 >= f1_best:
                f1_best = f1
                forest_best_depth = depth
                n_est_best = n_est
    
    #plt.plot(f1_score_list)
    #plt.title(f'Зависимость f1-score случайного леса от глубины \nпри количестве деревьев, равном {n_est_best}')
    #plt.xlabel('Глубина')
    #plt.ylabel('f1-score')
    #plt.show()
    
    # используя полученную глубину, ищем делаем тоже самое, но для деревьев 
    # относительно известной глубины и выводим зависимость на экран, 
    # а также запоминаем лучшую модель (по f1-score)
    f1_score_forest_best = 0
    ran_for_best = 'None'
    f1_score_list = []
    for trees in range(1,100):
        model = RandomForestClassifier(max_depth = forest_best_depth, 
                                    class_weight = 'balanced',
                                    n_estimators= trees, 
                                    random_state=12345)
        model.fit(features_train, target_train)
        f1 = f1_score(target_valid, model.predict(features_valid))
        f1_score_list.append(f1)
        if f1 >= f1_score_forest_best:
            f1_score_forest_best = f1
            n_estimators_best = trees
            ran_for_best = model
    plt.plot(f1_score_list)
    plt.title(f'Зависимость f1-score случайного леса от количества деревьев \nпри глубине {forest_best_depth}')
    plt.xlabel('Количество деревьев')
    plt.ylabel('f1-score')
    plt.show()
    print(f'Лучший f1-score для случайного леса: {f1_score_forest_best}\n\
        Лучшая глубина: {forest_best_depth}\n\
        Лучшее количество деревьев: {n_estimators_best}\n')
    # возвращаем лучшую по метрике f1-меры модель случайного леса и его основные метрики
    return ran_for_best, metrics(target_valid, 
        ran_for_best.predict(features_valid),
        ran_for_best.predict_proba(features_valid)[:,1],
        'RandomForestClassifier',
        type)
    
```

### Решающее дерево
Начнем с модели решающего дерева, подбрем лучшие гиперпараметры для достижения лучших результатов `f1-score` модели, а затем проверим полноту, точность и прочие метрики, а также посмотрим, как `f1-score` зависит от глубины (используем функцию, прописанную чуть выше, сохраним получившуюся лучшую модель решающего дерева в `DTC_best_disbalanced` (DTC = DecisionTreeClassifier), а метрики в наш словарь, который откроем после анализа всех трёх моделей для постановки выводов и прочего)

```python
DTC_best_disbalanced, all_metrics_disbalanced['DecisionTreeClassifier'] = decision_tree_research(features_train_scaled,
                                          target_train,
                                          features_valid_scaled,
                                          target_valid,
                                          type = 'disbalanced')
```

Ещё раз вспомним, какой у нас дисбаланс классов и запишем его в переменную `repeat `, округлив в большую сторону

```python
repeat = math.ceil(target.value_counts()[0]/target.value_counts()[1])
repeat
```

Сначала напишем функцию, с помощью которой будем получать необходимые выборки

```python
def upsample(features, target, repeat = 4): 
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    if len(target_zeros) > len(target_ones):
        features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
        target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
        features_upsampled = shuffle(features_upsampled, random_state = 12345)
        target_upsampled = shuffle(target_upsampled, random_state = 12345)
        return features_upsampled, target_upsampled
    else:
        features_upsampled = pd.concat([features_ones] + [features_zeros] * repeat)
        target_upsampled = pd.concat([target_ones] + [target_zeros] * repeat)
        features_upsampled = shuffle(features_upsampled, random_state = 12345)
        target_upsampled = shuffle(target_upsampled, random_state = 12345)
        return features_upsampled, target_upsampled
```

Сначала напишем функцию, с помощью которой будем получать необходимые выборки

```python
def downsample(features, target, fraction): 
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    if len(target_zeros) > len(target_ones):
        features_downsampled = pd.concat([features_zeros.sample(frac = fraction, 
                                                                random_state = 12345)] 
                                                                +[features_ones])
        target_downsampled = pd.concat([target_zeros.sample(frac = fraction, 
                                                            random_state = 12345)] 
                                                            + [target_ones])
        features_downsampled, target_downsampled = \
        shuffle(features_downsampled, target_downsampled, random_state = 12345)

        return features_downsampled, target_downsampled
    else:
        features_downsampled = pd.concat([features_ones.sample(frac = fraction, 
                                                                random_state = 12345)] 
                                                                +[features_zeros])
        target_downsampled = pd.concat([target_ones.sample(frac = fraction, 
                                                            random_state = 12345)] 
                                                            + [target_zeros])
        features_downsampled, target_downsampled = \
        shuffle(features_downsampled, target_downsampled, random_state = 12345)

        return features_downsampled, target_downsampled
```

Теперь определимся с тем, какую часть нам откидывать, попробуем добиться наилучшей балансировки классов
